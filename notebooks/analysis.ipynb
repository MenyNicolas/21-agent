```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../historique_BJ.csv")
```

```python
df.info()
```

```python
df["tc_arrondi"] = df["true_count"].round(0)  # ou .round(1) si tu veux plus fin
```

```python
esperance = df.groupby("tc_arrondi")["résultat_mise"].mean()
```

```python
esperance.plot(title="Espérance réelle par True Count", ylabel="Gain moyen (€ ou unités)")
```
