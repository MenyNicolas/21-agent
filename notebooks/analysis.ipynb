```python
import pandas as pd
```

```python
df = pd.read_csv("../historique_BJ.csv")
```

```python
df["résultat_mise"].sum()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

# Création de la colonne true_count_arrondi
df['true_count_arrondi'] = df['true_count'].round()

# Calcul de la somme des résultats par true_count_arrondi
resultat_par_true_count = df.groupby('true_count_arrondi')['résultat'].sum()

# Création du graphique
plt.figure(figsize=(12, 6))
resultat_par_true_count.plot(kind='bar')
plt.title('Somme des résultats par true_count_arrondi')
plt.xlabel('True Count Arrondi')
plt.ylabel('Somme des résultats')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
